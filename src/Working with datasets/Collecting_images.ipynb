## [Import libraries]()

```python
import pandas as pd
import numpy as np
import os
import cv2 as cv
import time
import uuid
```

## [Get the data from the webcam]()

```python
# config the path where the images will store in
path = "/home/hany_jr/Ai/FaceDetectionDeepLearning/Dataset/images"
uniqe_path = str(uuid.uuid1()) + ".jpg"
new_path = os.path.join(path, uniqe_path)
numberOfImages = 25
```

```python
# cofig camera

cap = cv.VideoCapture(0)

print("will start capturing images in 5 seconds")

time.sleep(5)

for image in range(numberOfImages):
    if not cap.isOpened():
        print("Error: Camera is not connected")
        break

    # get the current frame
    ret, frame = cap.read()

    if not ret:
        print("Error: Can't receive frame")
        break
    # generate uniqe name for each image
    uniqe_path = str(uuid.uuid1()) + ".jpg"
    new_path = os.path.join(path, uniqe_path)

    # save the image
    image = os.path.join(path, uniqe_path)
    cv.imwrite(image, frame)
    cv.imshow("frame", frame)
    time.sleep(1)

    if cv.waitKey(1) == ord("q"):
        break

cap.release()
cv.destroyAllWindows()
```
