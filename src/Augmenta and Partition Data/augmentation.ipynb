```python
import os
import sys
import numpy as np
import albumentations as albu
import cv2  # for image processing
import json  # for reading json files
```

```python
image = cv2.imread(
    "/home/hany_jr/Ai/FaceDetectionDeepLearning/Dataset/images/00a6fab8-7396-11ef-a98c-9c443d1ab0ba.jpg"
)


label = json.load(
    open(
        "/home/hany_jr/Ai/FaceDetectionDeepLearning/Dataset/labels/00a6fab8-7396-11ef-a98c-9c443d1ab0ba.json"
    )
)

box = label["shapes"][0]["points"]
cord1 = (int(box[0][0]), int(box[0][1]))
cord2 = (int(box[1][0]), int(box[1][1]))
new_imge = cv2.rectangle(image.copy(), cord2, cord1, (0, 255, 0), 2)

cv2.imshow("Original Image", new_imge)

if cv2.waitKey(0) == ord("q"):
    cv2.destroyAllWindows()
```

## [Create an augmentation pipline]()

```python
augmentations = albu.Compose(
    [
        # create a random crop
        albu.RandomCrop(500, 500),
        # flip the image horizontally
        albu.HorizontalFlip(p=0.5),
        # randomlly brighten the image
        albu.RandomBrightnessContrast(p=0.5),
        # add gamma contrast
        albu.RandomGamma(p=0.5),
        # shift rgb channels
        albu.RGBShift(p=0.5),
        # flip the image vertically
        albu.VerticalFlip(p=0.5),
    ],
    bbox_params=albu.BboxParams(format="pascal_voc", label_fields=["class_labels"]),
)

# bonderies = np.ndarray.flatten(np.array([image.shape[:2], image.shape[:2]])).tolist()

box = np.ndarray.flatten(np.array([cord2, cord1])).tolist()
box = [box[2], box[3], box[0], box[1]]
augmented = augmentations(image=image, bboxes=[box], class_labels=["face"])
```

## [Try augmentations pipline on a image]() 

```python
img = augmented["image"]

new_imge = img.copy()
if len(augmented["bboxes"]) > 0:
    boxes = augmented["bboxes"][0]

    cord1 = (int(boxes[0]), int(boxes[1]))
    cord2 = (int(boxes[2]), int(boxes[3]))
    new_imge = cv2.rectangle(img.copy(), cord1, cord2, (0, 255, 0), 2)

cv2.imshow("Augmented Image", new_imge)


if cv2.waitKey(0) == ord("q"):
    cv2.destroyAllWindows()
```

```python
# 1) test augmentation
test_images = os.listdir(
    "/home/hany_jr/Ai/FaceDetectionDeepLearning/Dataset/test/images"
)
test_anno = os.listdir(
    "/home/hany_jr/Ai/FaceDetectionDeepLearning/Dataset/test/annotations"
)
test_anno.sort()
test_images.sort()

for i in range(len(test_images)):
    if test_images[i].split(".")[0] != test_anno[i].split(".")[0]:
        print("Error in the test data")
        break
```

```python
# Augment the test data

test_url = "/home/hany_jr/Ai/FaceDetectionDeepLearning/Dataset/test/images"
test_anno_url = "/home/hany_jr/Ai/FaceDetectionDeepLearning/Dataset/test/annotations"


test_augm_url = "/home/hany_jr/Ai/FaceDetectionDeepLearning/Dataset/test_augm/images"
test_augm_anno_url = (
    "/home/hany_jr/Ai/FaceDetectionDeepLearning/Dataset/test_augm/labels"
)
for img_path in test_images:

    image = cv2.imread(os.path.join(test_url, img_path))
    label = {}

    anno_path = os.path.join(test_anno_url, img_path.split(".")[0] + ".json")

    # open json file to get the lavels
    with open(anno_path, "r") as f:
        label = json.load(f)

    # Get coordinates of the original image
    box = label["shapes"][0]["points"]
    cord1 = (int(box[0][0]), int(box[0][1]))
    cord2 = (int(box[1][0]), int(box[1][1]))

    for counter in range(60):
        flatten_box = np.ndarray.flatten(np.array([cord2, cord1])).tolist().copy()
        flatten_box = [
            flatten_box[2],
            flatten_box[3],
            flatten_box[0],
            flatten_box[1],
        ].copy()
        # cv2.imshow("original", image)
        augmented = augmentations(
            image=image.copy(), bboxes=[flatten_box], class_labels=["face"]
        )

        aug_img = augmented["image"].copy()

        new_imge = aug_img.copy()
        if len(augmented["bboxes"]) > 0:
            boxes = augmented["bboxes"][0]

            # add the augmented image to the directory
            # create a new json file for the augmented image with the new coordinates and the class label is face

            file = {
                "shapes": [
                    {
                        "label": "face",
                        "points": [
                            [boxes[2], boxes[3]],
                            [boxes[0], boxes[1]],
                        ],
                    }
                ],
            }
            cv2.imwrite(
                os.path.join(
                    test_augm_url, img_path.split(".")[0] + str(counter) + ".jpg"
                ),
                aug_img,
            )

            with open(
                os.path.join(
                    test_augm_anno_url, img_path.split(".")[0] + str(counter) + ".json"
                ),
                "w",
            ) as f:
                json.dump(file, f)
        else:
            file = {
                "shapes": [
                    {
                        "label": "no face",
                        "points": [
                            [-1, -1],
                            [-1, -1],
                        ],
                    }
                ],
            }
            cv2.imwrite(
                os.path.join(
                    test_augm_url, img_path.split(".")[0] + str(counter) + ".jpg"
                ),
                aug_img,
            )

            with open(
                os.path.join(
                    test_augm_anno_url, img_path.split(".")[0] + str(counter) + ".json"
                ),
                "w",
            ) as f:
                json.dump(file, f)
```
